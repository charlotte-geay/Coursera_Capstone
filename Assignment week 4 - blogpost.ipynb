# Personal project - "Applied data science capstone" course - Blogpost

This project shows and locate on a map the cities of a given French department that are near from a given venue. For example, the following map indicate the cities in which there is an historical site in the "Nord" department.

This map represents the cities in the "Nord" department of France for which there is at least one "Historic Site" located less than 1km from the center of the city.
